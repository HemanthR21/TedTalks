# **To split Date column into Month and Year**





```python
dt = []
month  = []
for value in df["date"]:
  dt.append(value.split(" ")[0])
  month.append(value.split(" ")[1])
print(dt)
print(month)

df["month"] = [x.split(" ")[0] for x in df["date"]]
```

```python
df["month"] = [x.split(" ")[0] for x in df["date"]]
df["year"] = [x.split(" ")[1] for x in df["date"]]

df.head()
```

```python
df = df.drop(["date"], axis = 1)

df.head()
```

```python
months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]

df["month"] = pd.Categorical(df['month'], categories=months, ordered=True)
```

```python
df.head()
```

```python
monthly_frequency = pd.DataFrame()
print(monthly_frequency)

monthly_frequency["count"] = df.groupby("month").size().sort_values(ascending = False)
```

# **Popular TED Talks by Months**

```python
import matplotlib.pyplot as plt
from datetime import datetime as dt
import numpy as np
import seaborn as sns
import plotly.express as px


fig = px.bar(monthly_frequency, x = monthly_frequency.index, y = "count", title = "Monthly Talks", color = 
             monthly_frequency.index,labels={'month':'Month', 'count': 'Frequency'})
fig.show()
```

```python
top_ted_talkers = {}

for row, value in df.iterrows():
  author = value["author"]

  if author not in top_ted_talkers:
    ted_talker = {}
    ted_talker["author_count"] = 1
    ted_talker["views"] = value["views"]
    ted_talker["likes"] = value["likes"]
    ted_talker["author"] = author
  else:
    ted_talker = top_ted_talkers.get(author)
    ted_talker["author_count"] += 1
    ted_talker["views"] += value["views"]
    ted_talker["likes"] += value["likes"]
    ted_talker["author"] = author
  
  top_ted_talkers[author] = ted_talker
```

# **Popular TED Talks by views generated**

```python
popular_talk_by_views = df.sort_values(by="views", ascending = False)


fig = px.bar(popular_talk_by_views[:15],x = "views",y="title",title='Most popular TED talks',
             orientation='h', color='title', width=1500, height=700, labels={'title':'TED talks title', 'views': 'Total number of views'})

fig.show()
```

# **Authour with highest views**

```python
ted_speakers  = pd.DataFrame(top_ted_talkers.values())

ted_speakers = ted_speakers.sort_values(by="author_count",ascending=False)


ted_speakers = ted_speakers[:15]

```

```python
ted_speakers

fig = px.treemap(ted_speakers, path = ["author"], values = 'author_count',hover_data = ['views','likes'])
fig.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
print(stopwords)
```

```python
comment_words = ''
for line in df.title:
    words = line.lower().split() #separar as palavras em uma lista#
    comment_words += " ".join(words)+" " 

wordcloud = WordCloud(width = 1000, height = 800, #customização do gráfico#
                background_color ='black',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)

plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 1)
 
plt.show()
```
